```python
# plot pdf and cdf
counts,bin_edges = np.histogram(df['HP'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(df['HP'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf)
plt.grid()
plt.show()

# It is observed that the speed is vary in the range of 7.5 to 25.
```

```python
# plot pdf and cdf
counts,bin_edges = np.histogram(df['MPG'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(df['MPG'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf)
plt.grid()
plt.show()

# It is observed that the speed is vary in the range of 7.5 to 25.
```

```python
#Skewness

from scipy.stats import skew
from scipy.stats import kurtosis
```

```python
sp_lst=df['HP'].tolist()
print(skew(sp_lst))
print(kurtosis(sp_lst,axis=0,bias=True))

#The skewness of the speed is positive to there is right skkew and data is distributed
```

```python
sp_lst=df['MPG'].tolist()
print(skew(sp_lst))
print(kurtosis(sp_lst,axis=0,bias=True))

#The skewness of the speed is negative to there is left skkew and data is distributed
```
