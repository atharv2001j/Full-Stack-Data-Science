## Step 7: PDF and CDF

```python
counts, bin_edges = np.histogram(df['speed'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(df['speed'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();

# It is observed that the speed is vary in the range of 7.5 to 25.
# 
```

## Step 8: Mean, Median, Mode, Standard Deviation, Variance

```python
x = np.mean(df['speed'])
y = np.median(df['speed'])
z = stats.mode(df['speed'])

print(f"Mean = {x}, Median = {y}, Mode = {z}")
```

```python
x = np.mean(df['dist'])
y = np.median(df['dist'])
z = stats.mode(df['dist'])

print(f"Mean = {x}, Median = {y}, Mode = {z}")
```

```python
std_speed = np.std(df['speed'])
std_dist = np.std(df['dist'])

var_speed = np.var(df['speed'])
var_dist = np.var(df['dist'])

print(f"Speed STD = {std_speed}, Speed VAR = {std_dist}")
print(f"Dist VAR = {var_speed}, Dist VAR = {var_dist}")
```

```python
sp_list = df['speed'].tolist()
print("Skewness of Speed", skew(sp_list))
```

```python
dist_list = df['dist'].tolist()
print("Skewness of Speed", skew(dist_list))
```

```python
print(kurtosis(df['speed'], axis = 0, bias = True))
```

```python
print(kurtosis(df['dist'], axis = 0, bias = True))
```
